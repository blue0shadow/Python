# Bokeh 한번에 제대로 배우기


## Bokeh 특징

* 최신 브라우저의 인터랙티브 시각화
* 독립형 HTML 문서 또는 서버 지원
* 표현력이 뛰어나고 다양한 그래픽 지원
* 큰 동적 데이터 또는 스트리밍 데이터 지원
* 파이썬(또는 Scala, R, ...)에서 쉽게 사용
* 자바스크립트 불필요

* 샘플 데이터 다운로드
