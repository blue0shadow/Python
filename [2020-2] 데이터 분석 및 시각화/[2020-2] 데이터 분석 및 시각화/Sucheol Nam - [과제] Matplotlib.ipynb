## 1. total_bill 과 tip 의 분포도(scatter) 시각화하기

### tip 금액에 따라 점의 크기 변화 주기, alpha 옵션으로 투명도 지정하기
### 축 label 추가하기, 플롯 title 추가하기


```python
x = A["tip"]
y = A["total_bill"]

sizes = 20 * A["tip"]

plt.scatter(x, y, s = sizes , alpha = 0.7)
plt.xlabel("tip")
plt.ylabel("total_bill")
plt.title("Scatter of total_bill and tip");
```

## 2. 성별에 따른 tip 비교하기 (bar plat)

```python
plt.bar(A["sex"], A["tip"],color = "royalblue", width  = 0.5)
plt.xlabel("sex", fontweight = "bold")
plt.ylabel("tip", fontweight = "bold")
plt.title("Comparision of tip with Sex");
```

## 3. 남자의 여자의 팁 금액의 분포도(histogram)를 그려 비교하세요

### 2 X 1 의 subplot을 이용하세요.

```python
x1 = A[A['sex'] == 'Male'].tip
x2 = A[A['sex'] == 'Female'].tip

d = dict( histtype = "stepfilled", alpha = 1, bins= 30)

plt.figure(figsize = (10,15))

plt.subplot(211)
plt.hist(x1, **d)
plt.title("Male_tip",pad = 10)
plt.xlabel("tip")
plt.ylabel("Frequency")
plt.grid();

plt.subplot(212)
plt.hist(x2, **d);
plt.xlabel("tip")
plt.ylabel("Frequency")
plt.title("Female_tip",pad = 10)

plt.grid();
```
