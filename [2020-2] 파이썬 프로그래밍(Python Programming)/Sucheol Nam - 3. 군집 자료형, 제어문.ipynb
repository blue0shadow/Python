### 3. 리스트를 원소로 갖는 리스트의 각 원소들의 합을 구하는 프로그램을 작성하시오.

a = [(1,2), (3,4),(6,7),(10,11)]

(hint) for (n1,n2) in a:



```
a = [(1,2), (3,4),(6,7),(10,11)]
sum = 0
for i in a:
  for j in i:
    sum += j

print(sum)
```

### 5. 2020년이 윤년인지 검사하는 프로그램을 작성하시오.


*  400 으로 나누어 떨어지는 해는 윤년이다.
*  4로 나누어 떨어지는 해는 윤년이다.
*  4로 나누어 떨어지지만 100으로 나누어 떨어지는 해는 윤년이 아니다.




```
a = int(input("연도를 입력하시오 :"))

if (a % 400 == 0) or ((a % 4 == 0) and (a % 100 != 0)):
    print("윤년이다.")
else:
    print("윤년이 아니다.")
```
