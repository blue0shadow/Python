### 1. 세 개의 양의 정수를 입력받아 그 합이 짝수이면 가장 큰 수를 출력하고, 홀수이면 가장 작은 수를 출력하는 프로그램을 작성하시오.

```
a = int(input("첫번쨰 양의 정수를 입력하세요. : "))
b = int(input("두번쨰 양의 정수를 입력하세요. : "))
c = int(input("세번쨰 양의 정수를 입력하세요. : "))

sum = (a + b + c)

if (sum % 2 )== 0:
  print("(짝수)가장 큰 수는 :", max(a,b,c))
else:
  print("(홀수)가장 작은 수는 :", min(a,b,c))
```

### 4. 사각형의 둘레를 구하는 함수를 작성하고, 가로 세로의 길이가 각각 3, 6인 사각형의 둘레를 함수를 이용하여 구해 보시오.

```
def square(x,y):
  sum = 2*(x+y)
  return sum

print(square(3,6))
```
